```python
from qiskit import QuantumRegister, QuantumCircuit 
```

```python
qr = QuantumRegister(2)
```

```python
circuit=QuantumCircuit(qr)
```

```python
circuit.cx(qr[0],qr[1])
circuit.cx(qr[1],qr[0])
circuit.cx(qr[0],qr[1])
```

```python
circuit.draw(initial_state = True, output='mpl')
```

```python
qr = QuantumRegister(3)
circuit = QuantumCircuit(qr)
circuit.h(qr[0])
circuit.cx(qr[0],qr[1])
circuit.cx(qr[1],qr[2])
circuit.draw(initial_state = True, output='mpl')
```

```python
qr=QuantumRegister(3)
circuit= QuantumCircuit(qr)
circuit.ry(1.9106, qr[0]) # where 1.9106 = 2*acos(1/sqrt(3))
circuit.ch(qr[0],qr[1])
circuit.cx(qr[1],qr[2])
circuit.cx(qr[0],qr[1])
circuit.x(qr[0])
```

```python
circuit.draw(initial_state = True, output='mpl')
```

```python
# Q.5(d)
```

```python
from qiskit import QuantumRegister, QuantumCircuit
qreg_q = QuantumRegister(3)
circuit = QuantumCircuit(qreg_q)

circuit.x(qreg_q[0])
circuit.x(qreg_q[1])
circuit.ccx(qreg_q[0], qreg_q[1], qreg_q[2])
circuit.x(qreg_q[0])
circuit.cx(qreg_q[1], qreg_q[0])
circuit.x(qreg_q[1])
circuit.swap(qreg_q[0], qreg_q[1])
circuit.cx(qreg_q[1], qreg_q[0])
#where number is given in the format of q[2]q[1]q[0]
```

```python
circuit.draw(output='mpl')
```
